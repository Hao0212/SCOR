```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import rgb2hex
%matplotlib inline
import seaborn as sns

#colormap and number
cmap = cm.get_cmap('GnBu',12) 
col_def =[]
for i in range(cmap.N):
    rgb = cmap(i)[:3]
    col_def.append(rgb2hex(rgb))
```

```python
# Import data from excel file
data = pd.read_excel('/Users/zhouhao/Documents/Datathon/Common_Repository/02_Data/2018/2018_Andhra Pradesh_Kharif.xlsx')
data
```

```python
data.isnull().sum()
# 2016 Yield has too much null value, Use the data from yield 2009 to 2015
```

```python
 # Use Average to fillna
data.iloc[:,20:27] = data.iloc[:,20:27].fillna(data.mean())
data.iloc[:,20:27].isnull().sum() 
```

```python
yield_mean_df = data.iloc[:,20:27].mean()
sns.set_style('darkgrid')
fig = plt.plot(figsize = (12,6))
fig = sns.barplot(x = yield_mean_df.values, y =yield_mean_df.index, edgecolor = "black", palette = col_def)
fig.set_title("Yield ranking (By Average)", fontsize = 20)
plt.show() # Show every year's average yield
```

```python
# Calculate the loss function
loss_df = data[['Sum Insured (Inr)','Indemnity Level','2009 Yield','2010 Yield','2011 Yield',
                '2012 Yield','2013 Yield','2014 Yield','2015 Yield']]
# Caculate the threshold
loss_function = []
for index, row in loss_df.iterrows():
    #filter the smallest two numbers
    filtered_series = row.iloc[2:].sort_values(ascending = True).iloc[2:]
    #
    y_bar = sum(filtered_series) * row[1] / 5
    value_list =[]
    for value in filtered_series:
        value_list.append(max(0,value - y_bar)/y_bar)
    loss_function.append(row[0] * sum(value_list))
loss_function
```
